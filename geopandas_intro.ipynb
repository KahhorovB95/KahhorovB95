# **Statik qiymat bilan yangi ustun yaratish**

# **Hisoblangan qiymatlar bilan yangi ustun yaratish**

Matn va raqamlar ustida turli xil ifodala qilsa bo'ladi

`layer[yangi_atribut_ustun] = layer[atribut1] + layer[atribut2]`

````python
layer['info'] = layer['Region_UZ'] + " viloyati"
layer
````

# **Shart asosida yangi ustun yaratish**



````python
layer.loc[layer['Region_UZ'] == "Qoraqalpog'iston Respublikasi", 'info'] = layer['Region_UZ']
layer.loc[layer['Region_UZ'] == "Toshkent ш.", 'info'] = 'Toshkent sh.'
layer
````

# **Geometrik qiymatga asoslangan ustun yaratish**

`layer['Area_km2'] = layer.geometry.area / 1e6`

````python
layer['x'] = layer.geometry.centroid.x
layer['y'] = layer.geometry.centroid.y
layer.head()
````
